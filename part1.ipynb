```python
import pandas as pd

# Cargar el conjunto de dates 
df = pd.read_csv("/Users/arigov/Downloads/COVID-19__Seq_enciaci__de_variants.csv")

# Visualizar les primeres files del conjunto de dades per entendre l'estructura
print(df.head())
```

```python
# Verificar la informació de les columnes i tipus de dades
print(df.info())
```

```python
# Obtener el número de files y columnes
num_filas, num_columnas = df.shape

# Imprimir el resultat
print(f"Número de filas: {num_filas}")
print(f"Número de columnas: {num_columnas}")

# Obtenerir la llista de columnes
columnas = df.columns.tolist()

# Imprimir la llista de columnes
print("Columnas del conjunto de datos:")
for columna in columnas:
    print(columna)
```

```python
for columna in df.columns:
    print(f"Valores únicos en la columna '{columna}':")
    print(df[columna].unique())
    print("-" * 50)
```

```python
print(df.describe())
```

```python
# Contar el número de caselles buides per columnes
recuento_celdas_vacias_por_columna = df.isnull().sum()

# Imprimir el recompte de caselles buides per columnes
print("Recuento de celdas vacías por columna:")
print(recuento_celdas_vacias_por_columna)
```

```python
import matplotlib.pyplot as plt

columnas_numericas = df.select_dtypes(include=['number'])

# Crear box-plots per cada columna numérica
for columna in columnas_numericas.columns:
    plt.boxplot(df[columna])
    plt.title(f'Box-plot para la variable {columna}')
    plt.show()
```
